```python
from sklearn.model_selection import train_test_split
```

```python
df_y=letter_df.loc[:,'letter']
```

```python
df_x=letter_df.loc[:,['xbox', 'ybox', 'width', 'height', 'onpix', 'xbar', 'ybar',
       'x2bar', 'y2bar', 'xybar', 'x2ybar', 'xy2bar', 'xedge', 'xedgey',
       'yedge', 'yedgex']]
```

```python
X_train, X_test, y_train, y_test = train_test_split(df_x, df_y, random_state = 47, test_size = 0.3)
```

```python
svc_linear = svm.SVC(kernel='linear', C=1,gamma='auto').fit(X_train, y_train)
```

```python
svc_rbf = svm.SVC(kernel='rbf', C=1,gamma='auto').fit(X_train, y_train)
```

```python
pred_linear=svc_linear.predict(X_test)
```

```python
pred_rbf=svc_rbf.predict(X_test)
```
